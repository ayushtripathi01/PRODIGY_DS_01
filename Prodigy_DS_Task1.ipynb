```python
# Example data (can be replaced with your actual dataset)
data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female', 'Other', 'Female']
}
df = pd.DataFrame(data)

# Count frequency of each category
gender_counts = df['Gender'].value_counts()

# Plot bar chart
plt.figure(figsize=(5, 3))
plt.bar(gender_counts.index, gender_counts.values, color='mediumturquoise', edgecolor='black')
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Gender Distribution")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Example data (can be replaced with your actual dataset)
data = {
    'Age': [18, 22, 19, 24, 30, 33, 45, 52, 60, 36, 29, 41, 23, 18]
}
df = pd.DataFrame(data)

# Create age groups
bins = [0, 18, 25, 35, 50, 100]
labels = ['0-18', '19-25', '26-35', '36-50', '51+']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)

# Count frequency of each group
age_group_counts = df['Age Group'].value_counts().sort_index()

# Plot bar chart
plt.figure(figsize=(6, 4))
plt.bar(age_group_counts.index.astype(str), age_group_counts.values, color='salmon', edgecolor='black')
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.title("Age Group Distribution")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
